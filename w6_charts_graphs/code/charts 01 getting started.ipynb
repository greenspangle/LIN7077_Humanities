# MatPlotLib
Get started by drawing some simple charts
## Import matplotlib and numpy

```python
import matplotlib.pyplot as plt
import numpy as np
```

## Create some 'x' and 'y' data
As much data or as little as is needed.

**NOTE:** For these simple examples it is not strictly necessary to create numpy arrays.
Python lists work equally well. Try it and see.

The reason that numpy is used in these examples is that when you search for notebooks online they will often be using numpy to analyse the data so it's important you can recognise numpy when you see it.

```python
xpoints = np.array([1, 8])
ypoints = np.array([3, 10])
```

## Generate and display the plot - example 1

```python
plt.plot(xpoints, ypoints)
plt.show()
```

## More (x, y) data points - example 2

```python
xpoints = np.array([1, 2, 6, 8])
ypoints = np.array([3, 8, 1, 10])

plt.plot(xpoints, ypoints)
plt.show()
```

## Styling the lines and points - example 3

```python
ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, marker='o', ms=20, mec='r', mfc='g')
plt.show()
```

## Multiple lines on the same plot - example 4

```python
y1points = np.array([3, 8, 1, 10, 5, 7])
y2points = np.array([8, 6, 4, 5, 3, 9])

plt.plot(y1points)
plt.plot(y2points)
plt.show()
```

## Multiple (x, y) lines on the same plot - example 5

```python
x1points = np.array([1, 2, 4, 5, 7, 8])
y1points = np.array([3, 8, 1, 10, 5, 7])


x2points = np.array([1, 2, 3, 6, 7, 8])
y2points = np.array([8, 6, 4, 5, 3, 9])

plt.plot(x1points, y1points)
plt.plot(x2points, y2points)
plt.show()
```

## Plotting polynomials

```python
fromto = 1.05
# xpoints = np.array(list(range(-top, top)))
xpoints = np.linspace(-fromto, fromto)

y1points = np.array([i for i in xpoints])
y2points = np.array([(i-1)*(i+1) for i in xpoints])
y3points = np.array([2*(i+1)*i*(i-1) for i in xpoints])
y4points = np.array([4*(i-1)*(i-1/3)*(i+1/3)*(i+1) for i in xpoints])
y5points = np.array([8*(i-1)*(i-1/2)*i*(i+1/2)*(i+1) for i in xpoints])

plt.plot(xpoints, y1points)
plt.plot(xpoints, y2points)
plt.plot(xpoints, y3points)
plt.plot(xpoints, y4points)
plt.plot(xpoints, y5points)
plt.show()
```

```python
# do the same again but this time just using python lists instead of numpy arrays to demonsrate that both work

fromto = 1.05
# xpoints = np.array(list(range(-top, top)))
xpoints = np.linspace(-fromto, fromto)

y1points = [i for i in xpoints]
y2points = [(i-1)*(i+1) for i in xpoints]
y3points = [2*(i+1)*i*(i-1) for i in xpoints]
y4points = [4*(i-1)*(i-1/3)*(i+1/3)*(i+1) for i in xpoints]
y5points = [8*(i-1)*(i-1/2)*i*(i+1/2)*(i+1) for i in xpoints]

plt.plot(xpoints, y1points)
plt.plot(xpoints, y2points)
plt.plot(xpoints, y3points)
plt.plot(xpoints, y4points)
plt.plot(xpoints, y5points)
plt.show()
```
