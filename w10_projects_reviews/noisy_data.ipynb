## noisy linear

```python
import matplotlib.pyplot as plt
import numpy as np
import random
```

```python
def f(x):
    return 2*x+1
def r(x):
    return random.gauss(0, 0.15) * x 
```

```python
num_points = 100
noise = 10

xline = np.array([0, 10])
yline = np.array([f(i) for i in xline])

xdata = [random.random() * noise for i in range(num_points)]
ydata = [f(i)+r(noise)           for i in xdata]


plt.plot(xdata, ydata, linestyle = 'none', marker = 'o', ms=2,  mec='orange') #mfc='orange')
plt.show()

plt.plot(xline, yline)
plt.plot(xdata, ydata, linestyle = 'none', marker = 'o', ms=2, mec='orange', mfc='orange')
plt.show()
```
