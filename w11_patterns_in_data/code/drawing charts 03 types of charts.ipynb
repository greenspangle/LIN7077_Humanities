```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.array(["Apples", "Bananas", "Cherries", "Damsons"])
y = np.array([3, 8, 1, 10])

plt.bar(x, y)
plt.show()
```

```python

x = np.array(["A", "B", "C", "D"])
y = np.array([3, 8, 1, 10])

plt.barh(x, y, color = "hotpink")
plt.show()
```

```python
x = np.random.normal(170, 10, 250)

plt.hist(x, color = "aqua")
plt.show() 
```

```python
y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Damsons"]
myexplode = [0.2, 0, 0, 0]

plt.pie(y, labels = mylabels, explode = myexplode, shadow = True)
plt.show() 
```
