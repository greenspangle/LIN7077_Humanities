## Adding titles labelling axes

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([200, 205, 210, 220, 230, 250, 270, 280, 300, 330])

font1 = {'family':'serif','color':'blue','size':20}
font2 = {'family':'serif','color':'darkred','size':15}

plt.title("Sports Watch Data", fontdict = font1, loc = 'left')
plt.xlabel("Average Pulse", fontdict = font2)
plt.ylabel("Calorie Burnage", fontdict = font2)

plt.plot(x, y)
plt.show()
```

## Grouping Charts into panels

```python
# top left
# re-use previous chart
plt.subplot(2, 2, 1)
plt.plot(x, y)


# top right
y2 = np.array([3, 8, 1, 10])
plt.subplot(2, 2, 2)
plt.plot(y2, 'd:c', ms = 10, mec = 'k', mfc = 'r', linewidth ='3')


# bottom left
z = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Damsons"]
plt.subplot(2, 2, 3)
plt.pie(z, labels = mylabels)


# bottom right
fromto = 1.05
xpoints = np.linspace(-fromto, fromto)

y1points = np.array([i for i in xpoints])
y2points = np.array([(i-1)*(i+1) for i in xpoints])
y3points = np.array([2*(i+1)*i*(i-1) for i in xpoints])
y4points = np.array([4*(i-1)*(i-1/3)*(i+1/3)*(i+1) for i in xpoints])
y5points = np.array([8*(i-1)*(i-1/2)*i*(i+1/2)*(i+1) for i in xpoints])

plt.subplot(2, 2, 4)
plt.plot(xpoints, y1points)
plt.plot(xpoints, y2points)
plt.plot(xpoints, y3points)
plt.plot(xpoints, y4points)
plt.plot(xpoints, y5points)


# and display all the charts
plt.show()
```
