# MatPlotLib
## Formating chart lines and points

```python
import matplotlib.pyplot as plt
import numpy as np
ypoints = np.array([3, 8, 1, 10])
```

## Line styles

```python
plt.plot(ypoints, linestyle = 'solid')
plt.show()
plt.plot(ypoints, linestyle = 'dotted')
plt.show()
plt.plot(ypoints, linestyle = 'dashed')
plt.show()
plt.plot(ypoints, linestyle = 'dashdot')
plt.show()
plt.plot(ypoints, linestyle = 'none')
plt.show()
```

## Point styles

```python
plt.plot(ypoints, marker = 'o')
plt.show()
plt.plot(ypoints, marker = 's')
plt.show()
plt.plot(ypoints, marker = 'X')
plt.show()
plt.plot(ypoints, marker = 'D')
plt.show()
```

## Style format strings

```python
plt.plot(ypoints, 's--b', ms= 10, mec='m', mfc='g')
plt.show()

plt.plot(ypoints, 'd:c', ms = 20, mec = 'k', mfc = 'r', linewidth ='5')
plt.show()
```
